```python
mean = np.array([0, 0 ]).T
covariance_1 = np.array([[1, 0], [0, 1]])
covariance_2 = np.array([[2, 1], [1, 2]])
two_dimensional_normal_distribution_1 = np.random.multivariate_normal(mean, covariance_1, 5000)
```

```python
two_dimensional_normal_distribution_2 = np.random.multivariate_normal(mean, covariance_2, 5000)
```

## 5.

```python
alpha = np.random.normal(2, 0.1, 1000)
betha = np.random.normal(0.2, 0.2, 1000)
x = np.random.normal(0, 3, 1000)
sigma = abs(np.random.normal(0, 1, 1000))
mi = alpha + betha * x
y = np.random.normal(mi, sigma)
```

## 6.

```python
plt.scatter(x, y)
x_axis = np.linspace(-6, 6, 1000)
for num in range(1000):
    plt.plot(x_axis, alpha[num] + betha[num] * x_axis, "green")
plt.show()
```
